```python
#from google.colab import files

df = pd.read_csv("Telco.csv")

#uploaded = files.upload()
```

```python
#import io

#df = pd.read_csv(io.BytesIO(uploaded['Churn.csv']))
print(df)
```

```python
df = df.drop(columns=['customerID'])
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
X = df[["MonthlyCharges", "TotalCharges"]]
Y = df["Churn"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.8, random_state=1)

numNeighbors = [1, 3, 5, 11, 15, 25, 35, 45, 55]
trainAcc = []
testAcc = []

for k in numNeighbors:
  clf = KNeighborsClassifier(n_neighbors=k, metric='euclidean', p=2)
  clf.fit(X_train, Y_train)
  Y_predTrain = clf.predict(X_train)
  Y_predTest = clf.predict(X_test)
  trainAcc.append(metrics.accuracy_score(Y_train, Y_predTrain))
  testAcc.append(metrics.accuracy_score(Y_test, Y_predTest))

plt.plot(numNeighbors, trainAcc, 'ro-', numNeighbors, testAcc,'bv--')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
```
